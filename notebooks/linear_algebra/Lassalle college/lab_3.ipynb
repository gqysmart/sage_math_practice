```python
#exe 3.1
A= np.array([[2,1],
            [1,-3]])
b = np.array([[7],[-7]])
x = np.linalg.solve(A,b)
print("la solution de Ax=b est", x)

b_cap = A @ x
print("la vérification de Ax=b donne b_cap=", b_cap)
```

```python
# exe 3.2
A_inv = np.linalg.inv(A)
print("l'inverse de A est\n", A_inv)
x_2 = A_inv @ b
print("la solution de Ax=b en utilisant l'inverse de A est", x_2)
print("la methode de solve utilse la methode de gauss elimination pour trouver la solution de Ax=b,\n tandis que l'utilisation de l'inverse de A est une méthode plus générale mais moins efficace numériquement.")
```

```python
# exe 4.1
C = np.array([[4,1],
              [1,4]])
lamdas,v = np.linalg.eig(C)
print("les valeurs propres de C sont", lamdas)
print("les vecteurs propres de C sont\n", v)
```

```python
# exe 4.2
v1 = v[:,0]
print("le premier vecteur propre est", v1)
print("la shape de v1 est", v1.shape)
lambda_1 = lamdas[0]
print ("la première valeur propre est", lambda_1)
Cv1 = C@ v1
print("le produit de C et v1 est", Cv1)
lambda1_v1 = lambda_1 * v1
print("le produit de la première valeur propre et v1 est", lambda1_v1)

print("On peut voir que Cv1 et lambda1_v1 sont égaux, \nce qui confirme la définition des valeurs propres et des vecteurs propres.")
```
