Réponse 1: 
Le raisonnement probabliste en IA consiste à utiliser les probabiltés pour modeliser et gérer l'incertitude dans les situations où l'information est incomplète ou incertaine.
- Les probabilités permettent de représenter cette incertitude en attributant un degré de croyance à différents événements ou hypothèses, plutôt que de prendre des décisions strictement vraies ou fausses.

- Une probabilité a priori représente la croyance initiale sur un événement avant l'observation de nouvelles données, tandis que une probabilité a posteriori est une croyance mise à jour après avoir pris en compte ces nouvelles informaitons, souvent à l'aide du théorème de Bayes.

- Le diagnostic médical assisté par l'IA: avant un test, un système peut estimer la probabilité qu'un patient ait une maladie(a priori), puis adjuster cette probabilité après les résultas du test (a posteriori)


Réponse 2:
L'apprentissage profond est une branche de l'apprentissage automatique qui utilise des réseaux neurones comportant plusieurs couches cachés afin de d'extraire automatiquement des représentations complexes des données.

- Contrairement à un réseau de neuronnes traditionnel, qui possède peu de couches et repose souvent sur des caractéristiques définies manuellement, l'apprentissage profond permet d'apprendre des caractéristiques hiérarchiques de manière autonome à partir de données brutes.

- La reconnaissance d'images, comme la détection de visages, où plusieurs couches sont nécessaires pour identifier successivement les contours, les formes et les objects complexes.

## Section 3 Mise en situation

![section 3](./images/01_section3_1.png)

**Situation 1:**
1. Un apprentissage supervisé serait approprié, avec un réseau de neurones convolutionnel(CNN), car le diagnostic repose sur des images étiquettées, et les CNN sont particulièrement performants pour l'analyse visuelle et la détection de motifs complexes dans les images médicales.

1. Normalisation de la taille et de la résolution des images; Amélioration du contraste et réduction du bruit; Suppression des éléments non pertinents; Augmentation des données pour améliorer le robustesse du modèle.

1. Le risque de faux positifs ou faux négatifs, pouvant entaîner de l'anxiété ou un retard de diagnostic; La confidentialité des données médicales, qui sont extrêmement sensibles.

1. La fiabilité peut être assurée par une mise à jour régulière du modèle, une validation continue sur de nouvelles données cliniques; Et une collaboration avec des professionnels de la santé pour évaluer les performances réelles.

**Situation 2:**
1. Un apprentissage supervisé, combiné à des modèles temporels, est adapté, car le système apprend à prédire les congestions à partir de données historiques et en temps réel associées à des situations connues.

1. Données structurées: données GPS(vitesse, position, temps), capteurs de trafic ;Données non structurées: images ou vidéos provenant des caméras urbaines.

1. Risque technique: erreurs de prédiction, panne du système ou données incomplètes; Risque sociaux: surveillance excessive, atteinte à la vie privée, exclusion de certains quartiers.

1. Il est essentiel d'évaluer régulièrement les performances du système dans différents quartiers, d'analyser les biais potentiels des données; Et impliquer des experts en mobilité et des acteurs publics pour garantir une prise de décision équitable.
